### What countries had the highest Paid Maternal Leave in 2018?

```python
def plot_barh(metric, title, ylabel):
    plt.figure(figsize=(12,10))
    plt.title(title)
    plt.xlabel(ylabel)
    plt.barh(df['country_name'], df[metric]);
    return None
```

```python
plot_barh('pml_weeks', 'Paid Maternity Leave in Weeks by Country (2018)', 'Paid Maternity Leave (weeks)')
plt.savefig('../images/fig1.jpg', dpi=200)
```

### What countries have the highest PML average pay rate in 2018?

### What countries are best at matching the PML full pay equivalent to the total PML in 2018?

```python
def compare_barh(metric1, metric2, label1, label2, title, xlabel):
    plt.figure(figsize=(12,10))
    ax1 = plt.subplot(1,1,1)
    w = 0.3
    metric1_plot = ax1.barh(df['country_name'], df[metric1], color='b', align='center')
    ax2 = ax1.twinx()
    metric2_plot = ax2.barh(df['country_name'], df[metric2], color='g',align='center')
    plt.legend([metric1_plot, metric2_plot],[label1, label2])
    plt.title(title)
    plt.xlabel(xlabel);
    return None
```

```python
metric1 = 'pml_weeks'
metric2 = 'pml_full_equiv_weeks'
label1 = 'Paid Maternity Leave (weeks)'
label2 = 'PML Full Equivalent (weeks)'
title = 'What countries are best at matching the PML full pay equivalent to the total PML in 2018?'
xlabel = 'Number of Weeks'
compare_barh(metric1, metric2, label1, label2, title, xlabel)
plt.savefig('../images/fig3.jpg', dpi=200)
```

### How has the total number of countries administering 100% of maternity leave benefits changed over the years? 

##### Maternity leave is for biological mothers only. You take it while you are pregnant and after the baby is born. You can take a parental leave after your maternity leave. Paternity leave is for biological fathers only.

```python
temp = df[df['indicator_code'] == 'SH.MMR.LEVE.GT'].sum()
temp = pd.DataFrame(temp, index=temp.index, columns=['sum'])
temp['year'] = temp.index
temp.reset_index(drop=True, inplace=True)
temp.drop(index=[0], inplace=True)
temp.drop(index=[1], inplace=True)
temp.drop(index=[2], inplace=True)
temp.reset_index(drop=True, inplace=True)
temp['year'] = temp['year'].map(lambda x: int(x))
temp['sum'] = temp['sum'].map(lambda x: int(x))
```

```python
plt.figure(figsize=(12,10))
plt.title('Total Number of Countries administering 100% of Maternity Leave Benefits by Year', size=15, pad=20)
ax = sns.barplot(x=temp['sum'], y=temp['year'], orient='h');
ax.set(xlabel='Number of Countries', ylabel='Year');
plt.savefig('../images/fig6.jpg', dpi=200)
```

### How does Length of Paid Maternity Leave (weeks) compare across countries in 2019?

```python
temp = df[df['indicator_code'] == 'SH.MMR.LEVE'][['country_name', '2019']]
temp['2019'] = temp['2019'].map(lambda x: x // 7)
temp.dropna(inplace=True)
temp.reset_index(drop=True, inplace=True)
temp = temp.sort_values(by='2019', ascending=False).head(40).sort_values(by='2019')
```

```python
fig, ax = plt.subplots(figsize=(12,11))
plt.title('Top 40 Countries offering longest Paid Maternity Leave (weeks) in 2019', \
            size=15, pad=20, color='black', fontweight='bold')
plt.xlabel('Paid Maternity Leave (weeks)', labelpad=10)
ax.barh(temp['country_name'], temp['2019'], color='teal');
x = temp['2019']
for i, v in enumerate(x):
    ax.text(v, i-0.3, str(v), color='black', fontweight='normal')
plt.savefig('../images/fig7.jpg', dpi=200)
```

### How does women's employment rate change depending on education?

```python
df = pd.read_csv('../data/employment_by_children_1564_edu_level.csv')
```

```python
def plot_by_education_level(country):
    df1 = df[df['country'] == country]
    df1 = df1.T.reset_index()
    df1.columns = df1.iloc[1].values
    df1.drop(index=[0], inplace=True)
    df1.drop(index=[1], inplace=True)
    df1.rename(columns={'age' : 'year'}, inplace=True)
    df1 = df1.reset_index()
    df1.drop(columns=['index'], inplace=True)

    plt.figure(figsize=(12,5))
    ax1 = plt.subplot(1,1,1)
    w = 0.3
    high_edu = ax1.bar(df1['year'], df1['-High education'], color='b', align='center')
    ax2 = ax1.twiny()
    ax3 = ax1.twiny()
    med_edu = ax2.bar(df1['year'], df1['-Medium education'], color='g',align='center')
    low_edu = ax3.bar(df1['year'], df1['-Low education'], color='y',align='center')
    plt.legend([low_edu, med_edu, high_edu],['Low', 'Med', 'High'], loc='upper left')
    plt.title(country + ' Employment Rate by Year and Education Level', pad=60, fontdict = {'fontsize' : 20});
    plt.tight_layout()
    plt.savefig('../images/'+country+'.jpg', dpi=200);
    return None
```

```python
df.replace('..', 0, inplace=True)
```

```python
df.replace('No children aged 0-17', 'No children aged 0-14', inplace=True)
df.replace('At least one child aged 0-17', 'At least one child aged 0-14', inplace=True)
df.replace('No children aged 0-15', 'No children aged 0-14', inplace=True)
df.replace('At least one child aged 0-15', 'At least one child aged 0-14', inplace=True)
```

```python
df[['1998', '1999', '2000', '2001', '2002', '2003',
    '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012',
    '2013', '2014']] \
    = df[['1998', '1999', '2000', '2001', '2002', '2003',
          '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012',
          '2013', '2014']].astype(float)
```

```python
plot_by_education_level('Romania')
```

```python
plot_by_education_level('United States')
```

```python
plot_by_education_level('France')
```

### What happens with the employment rate for women in "country" between 1998 and 2014?

```python
def get_data(country):
    y = df[(df['country'] == country)] \
        [['1998', '1999', '2000', '2001', '2002', '2003',
          '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012',
          '2013', '2014']]
    x = ['1998', '1999', '2000', '2001', '2002', '2003',
         '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012',
         '2013', '2014']
    y = y.values.ravel()
    x = [int(item) for item in x]
    y = np.array(y)
    y = y.reshape(5,17)
    columns = df[(df['country'] == country)]['age'].values
    dates = x
    values = y.T
    data = pd.DataFrame(values, dates, columns=columns)
    return data
```

```python
country = 'United Kingdom'
data = get_data(country)
plt.figure(figsize = (12,8))
plt.title(country + ' Employment Rate by Education Level and Number of Children (1998-2014)')
sns_plot = sns.lineplot(data=data, palette="tab10", linewidth=2.5);
plt.tight_layout()
plt.savefig('../images/fig9.jpg', dpi=200);
```

```python
country = 'United States'
data = get_data(country)
plt.figure(figsize = (12,8))
plt.title(country + ' Employment Rate by Education Level and Number of Children (1998-2014)')
sns_plot = sns.lineplot(data=data, palette="tab10", linewidth=2.5);
plt.tight_layout()
plt.savefig('../images/fig10.jpg', dpi=200);
```
